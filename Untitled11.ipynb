```python
import pandas as pd
import numpy as np #imports
```

```python
file = "./pokemon.csv" #file path
```

```python
df_original = pd.read_csv(file)
df_original.head(20) #display file
```

```python
df_special = df_original[["Name", "Type 1", "HP","Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
df_special
```

```python
df_grouped = df_special.groupby(["Type 1"])
df_grouped.mean()
```

```python
totalp_df = df_grouped.sum()
totalp_df
```

```python
total_df = df_grouped["HP"] + df_grouped["Attack"] + df_grouped["Defense"] + df_grouped["Sp. Atk"] + df_grouped["Sp. Def"] +df_grouped["Speed"]
total_df
```

```python
pokedex_df = totalp_df.loc[totalp_df["Total Power"]]
pokedex_df
```
